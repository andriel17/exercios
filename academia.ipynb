```python
import pandas as pd
import numpy as np

# Criar um dataset fict√≠cio
data = {
    'MemberID': np.arange(1, 101),
    'Age': np.random.randint(18, 65, size=100),
    'Gender': np.random.choice(['Male', 'Female'], size=100),
    'ExerciseType': np.random.choice(['Cardio', 'Weightlifting', 'Yoga', 'Pilates'], size=100),
    'Duration': np.random.randint(30, 90, size=100),
    'Date': pd.date_range(start='2024-01-01', periods=100, freq='D')
}

df = pd.DataFrame(data)

# Exibir as primeiras linhas do dataset
print(df.head())

```

```python
import matplotlib.pyplot as plt

# Calcular a dura√ß√£o m√©dia por tipo de exerc√≠cio
avg_duration = df.groupby('ExerciseType')['Duration'].mean()

# Criar o gr√°fico
plt.figure(figsize=(10, 6))
avg_duration.plot(kind='bar', color='skyblue')
plt.title('Dura√ß√£o M√©dia do Exerc√≠cio por Tipo de Exerc√≠cio')
plt.xlabel('Tipo de Exerc√≠cio')
plt.ylabel('Dura√ß√£o M√©dia (min)')
plt.grid(axis='y')

# Exibir o gr√°fico
plt.show()
```
